```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
home_df = pd.read_csv("./homeprices.csv")
home_df
```

```python
# visualizing the dataset, in order to figure out what learning model to use
plt.scatter(home_df.area,home_df.price,marker="x")
plt.xlabel("area")
plt.ylabel("price")
plt.plot()
```

```python
plt.scatter(home_df.age,home_df.price,marker="x")
plt.xlabel("age")
plt.ylabel("price")
plt.plot()
```

```python
home_df.bedrooms.fillna(home_df.bedrooms.median(),inplace=True)
home_df
```

```python
plt.scatter(home_df.bedrooms,home_df.price,marker="x")
plt.xlabel("bedrooms")
plt.ylabel("price")
plt.plot()
```

```python
## starting model trading
X_train = home_df.drop("price",axis=1)
y_train = home_df.price

X_train=np.array(X_train)
y_train = np.array(y_train)

reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
reg.score(X_train,y_train)
```

```python
w = reg.coef_
b = reg.intercept_

print(w)
print(b)
```

```python
def model(params):
    X = np.array(params)
    return np.dot(w,X)+b
```

```python
options= [2500,4,5]
print(reg.predict([options]))
print(model(options))
```
