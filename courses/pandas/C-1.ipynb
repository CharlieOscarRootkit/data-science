## *`Maps`*

```python
# map(), transform the values of a Series, and returns a new Series, with transformed values
reviews
```

```python
alt_price = reviews.price.map(lambda x : 0  if pd.isna(x)  else x*5)
alt_price
```

```python
# return the index of the occurance of the maximum value, in a series
alt_price.idxmax()
```

```python
print(alt_price.loc[alt_price.idxmax()])
print(alt_price.max())
```

```python
# Create a DataFrame
data = {'A': [10, 20, 15], 'B': [25, 30, 20]}
df = pd.DataFrame(data)

# Get the index of the maximum value along each column
max_index_column = df.idxmax(axis=0)

print("Index of maximum value along columns:")
max_index_column
```
